```python
'''
P1: Test a Perceptual Phenomenon
    
1. What is our independent variable? What is our dependent variable?

ANSWER:
Independent Variable: In this case our independent variables are: Color of the text, Format of the text/symbols 
which are colored.
Dependent Variable: The time(in seconds) it takes for test takers to identify the color of the text or symbol.

2. What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? 
Justify your choices.

ANSWER:
μC and μI, respectively the congruent and incongruent population means.
H0: μC = μI
    The change in the time taken by the participants in the congruent and incongruent test is purely by chance 
    and is effectively not influence by congruence of the color.
    
H1: μC != μI
    The change in the time taken by the participants in the congruent and incongruent test is not due to chance 
    and is effectively influenced by the congruence of the color.
   
Statistical test:
I will be using the Student's t-test as we are trying to find if the difference in the mean times of the congruent 
and incongruent tests is by chance or not. We are making the assumptions that the samples are from a population that 
is normally distributed and that their variances are equal(if not it would have to be the Welch's t-test.)

Sources:
http://www.biostathandbook.com/testchoice.html
http://www.biostathandbook.com/variabletypes.html
'''
```

```python
'''
4. Provide one or two visualizations that show the distribution of the sample data. 
Write one or two sentences noting what you observe about the plot or plots.

ANSWER:
Based on the analysis below we can see that there is a higher proportion of people who took a longer amount of time
in the incongruent test as compared to the congruent test.

The hisograms show us that while the amount of time taken for the Congruent test ranges from approximately 8 to 22,
for the Incongruent test it ranges from approximately 15 to 27 with a couple of outliers at approximately 35.

From the plots generated we can roughly assume that it took longer for the Incongruent 
reposondents than the Congruent respondents.
'''
import matplotlib.pyplot as plt
%pylab inline
plt.scatter(stroopdata['Congruent'], stroopdata['Incongruent'], color = ['red','green'])
plt.xlabel('Number of respondents')
plt.ylabel('Number of seconds')
plt.show()
plt.hist(stroopdata['Congruent'], color = 'green')
plt.title('CONGRUENT')
plt.xlabel('Number of seconds')
plt.ylabel('Number of respondents')
plt.show()
plt.hist(stroopdata['Incongruent'], color = 'red')
plt.title('INCONGRUENT')
plt.xlabel('Number of seconds')
plt.ylabel('Number of respondents')
plt.show()
```

```python
'''
5. Now, perform the statistical test and report your results. What is your confidence level and your 
critical statistic value? Do you reject the null hypothesis or fail to reject it? 
Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

ANSWER:

I am performing the Students t-test with a confidence level of 95%. With the p value much lower than 0.05
we can reject our Null Hypothesis and state that the time taken in the incongruent test is more than
the congruent test in a statistically significant kind of way. Yes the results did match my expectations as 
I had tried experiment on myself and I too took longer in the Incongruent test.

Source: http://docs.scipy.org/doc/scipy-0.17.0/reference/generated/scipy.stats.kruskal.html
'''
from scipy import stats
x = stats.ttest_rel(stroopdata['Congruent'], stroopdata['Incongruent'])
print('H Statistic = ',x[0])
print('P Value = ',x[1])
```
